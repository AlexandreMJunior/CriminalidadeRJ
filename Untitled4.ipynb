```python
from tabulate import tabulate
from termcolor import colored

municipios = ['Rio de Janeiro', 'Nova Iguaçu', 'Seropédica', 'Belford Roxo', 'Duque de Caxias',
              'São Gonçalo', 'Niteroi', 'Petrópolis', 'Teresópolis', 'Macaé', 'Arraial do cabo',
              'Nova friburgo', 'Angra dos Reis']

for municipio in municipios:
    data_regiao = data[data['munic'] == municipio]

    # Calcular a soma de cada tipo de crime para essa região
    crimes = data.columns[9:-1]  # Selecionar todas as colunas de crimes
    somas_crimes = data_regiao[crimes].sum()  # Somar os crimes para a região

    # Ordenar as somas em ordem decrescente
    #somas_crimes = somas_crimes.sort_values(ascending=False)

    # Selecionar os 5 crimes mais frequentes
    top_5_crimes = somas_crimes[:5]

    # Criar uma lista de tuplas contendo os dados dos crimes
    dados_crimes = [(crime, quantidade) for crime, quantidade in top_5_crimes.items()]

    # Exibir a tabela formatada com cores
    print(f"\nOs 5 crimes mais frequentes em {municipio}:")
    tabela_formatada = tabulate(dados_crimes, headers=["Tipo de Crime", "Quantidade"], tablefmt="fancy_grid")
    tabela_colorida = '\n'.join([colored(line, color='cyan') for line in tabela_formatada.split('\n')])
    print(tabela_colorida)
```
